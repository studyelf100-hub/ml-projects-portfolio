# 02 - Model Training

Training three different models on each dataset separately.

```python
import sys
sys.path.append("../src")
import pandas as pd
from preprocess import load_and_prepare
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings("ignore")
```

## Load Data

```python
X_ng, y_ng, classes = load_and_prepare("../data/nigeria_students.csv")
X_pt, y_pt, _ = load_and_prepare("../data/portugal_students.csv")

X_train_ng, X_test_ng, y_train_ng, y_test_ng = train_test_split(X_ng, y_ng, test_size=0.2, random_state=42)
X_train_pt, X_test_pt, y_train_pt, y_test_pt = train_test_split(X_pt, y_pt, test_size=0.2, random_state=42)

print("Train/test split done.")
print(f"Nigeria - Train: {len(X_train_ng)}, Test: {len(X_test_ng)}")
print(f"Portugal - Train: {len(X_train_pt)}, Test: {len(X_test_pt)}")
```

## Train Models on Nigeria Data

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42)
}

ng_results = {}
for name, model in models.items():
    model.fit(X_train_ng, y_train_ng)
    score = model.score(X_test_ng, y_test_ng)
    ng_results[name] = score
    print(f"{name}: {score:.3f}")

print("\nBest model on Nigeria data:", max(ng_results, key=ng_results.get))
```

## Train Models on Portugal Data

```python
pt_results = {}
pt_models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42)
}

for name, model in pt_models.items():
    model.fit(X_train_pt, y_train_pt)
    score = model.score(X_test_pt, y_test_pt)
    pt_results[name] = score
    print(f"{name}: {score:.3f}")

print("\nBest model on Portugal data:", max(pt_results, key=pt_results.get))
```

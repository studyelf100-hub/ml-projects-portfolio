# 01 - Data Exploration

Let's understand what we're working with before touching any models.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ng = pd.read_csv("../data/nigeria_students.csv")
pt = pd.read_csv("../data/portugal_students.csv")

print("Nigeria dataset shape:", ng.shape)
print("Portugal dataset shape:", pt.shape)
```

## Grade Distribution by System

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

ng["final_grade"].value_counts().plot(kind="bar", ax=axes[0], color="#2E86AB", title="Nigeria - Grade Distribution")
pt["final_grade"].value_counts().plot(kind="bar", ax=axes[1], color="#A23B72", title="Portugal - Grade Distribution")

axes[0].set_xlabel("Grade")
axes[1].set_xlabel("Grade")
plt.tight_layout()
plt.savefig("../results/figures/grade_distribution.png", dpi=150)
plt.show()
```

## Feature Correlations

```python
numeric_cols = [
    "study_hours_per_week", "attendance_rate", "parent_education_level",
    "has_internet", "previous_score", "teacher_student_ratio"
]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.heatmap(ng[numeric_cols].corr(), annot=True, fmt=".2f", ax=axes[0], cmap="coolwarm", center=0)
axes[0].set_title("Nigeria - Feature Correlations")

sns.heatmap(pt[numeric_cols].corr(), annot=True, fmt=".2f", ax=axes[1], cmap="coolwarm", center=0)
axes[1].set_title("Portugal - Feature Correlations")

plt.tight_layout()
plt.savefig("../results/figures/feature_correlations.png", dpi=150)
plt.show()
```

## Key Observations

- Study hours and attendance are positively correlated with performance in both systems
- Teacher-student ratio is a more critical factor in Nigeria data
- Internet access is nearly universal in Portugal, making it a weaker discriminating feature there
